```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data_frame=pd.read_csv("C:\\Users\\Admin\\Downloads\\insurance_data.csv")
data_frame
```

```python
plt.xlabel("Age") 
plt.ylabel("Prob to buy insurrance") 
plt.scatter(data_frame.age,data_frame.bought_insurance,marker='+',color='red')
```

```python
data_frame.shape #it will tell me have how many no. of rows and columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data_frame[['age']]
X
```

```python
y=data_frame['bought_insurance']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.9,random_state=10)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
insurance_model=LogisticRegression() #Creating object of the model
```

```python
insurance_model.fit(X_train,y_train)
```

```python
X_test
```

```python
insurance_model.predict(X_test)
```

```python
insurance_model.predict([[80]])
```

```python
insurance_model.score(X_test,y_test)
```

```python
import pickle
```

```python
with open('insurance_model.pkl', 'wb') as file:
    pickle.dump(insurance_model, file)
```

```python
with open('insurance_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
prediction = loaded_model.predict(X_test)
prediction
```
