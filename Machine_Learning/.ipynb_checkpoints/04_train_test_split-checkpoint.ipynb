```python
#We are going to implememt Train Test Split Method. In this mthod we will divide our dataset into 80:20
#we will use 80% od data to train our model and rest 20% to test our model accuracy

import pandas as pd
```

```python
data_frame=pd.read_csv("C:\\Users\\Admin\\Downloads\\carprices.csv")
data_frame
```

```python
from sklearn.linear_model import LinearRegression 
car_model=LinearRegression()
```

```python
X=data_frame.drop(['Sell Price($)'],axis=1)
X
```

```python
Y=data_frame[['Sell Price($)']]
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Now we will be using the train_test_split() method to define the percentage of data we are using for training and testing.
# This method returns four parameters: X_train and X_test (feature datasets) & y_train and y_test (target datasets).
# We will specify test_size=0.2 to use 20% of the data for testing and 80% for training.
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state=42)

#random_state=42 will ensure we get same data for tarining and testing every time we run the code
```

```python
len(X_train)
```

```python
#It will choose random sample by default
X_train
```

```python
from sklearn.linear_model import LinearRegression
car_price_model=LinearRegression()
```

```python
car_price_model.fit(X_train,Y_train)
```

```python
X_test
```

```python
car_price_model.predict(X_test)
```

```python
car_price_model.score(X_test,Y_test)
```

```python
Y_test
```
