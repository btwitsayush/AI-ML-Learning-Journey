```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # this library is used for isual representation of data like graphs and charts
from sklearn import linear_model #it will help us to make prediction based on linear relationships b/w variables
```

```python
dataFrame=pd.read_csv("C:\\Users\\Admin\\Downloads\\prediction_sheet.csv")
dataFrame
```

```python
#it will allow us to open plot within our notebook
%matplotlib inline  
plt.xlabel('area(sq.ft')
plt.ylabel('Price(Rs')
plt.scatter(dataFrame.Area,dataFrame.Price,color='red',marker='+') #allow us to create scatter plot b/w 2 variables
```

```python
linear_regression=linear_model.LinearRegression()
# Fit the model to the training data
# The first argument must be a 2-D array (here, we reshape the Area column)
# The second argument is the target variable (Price)
linear_regression.fit(dataFrame[['Area']], dataFrame.Price)
```

```python
# We are calcullating this uisng linear equation formula  y=mx+b where m=slope,b= Y Intercept
linear_regression.predict([[3300]])
```

```python
linear_regression.coef_ #it will tell value of m
```

```python
linear_regression.intercept_ #it will tell value of y
```
